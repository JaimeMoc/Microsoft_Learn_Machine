```python
# Importar la biblioteca pandas con el alias pd
import pandas as pd

# Descargar archivos desde GitHub utilizando wget
!wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/graphing.py
!wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/doggy-boot-harness.csv
# Instalar la biblioteca statsmodels utilizando pip
!pip install statsmodels

# Crear un diccionario con los datos
datos = {
    "tamaño_arranque" : [ 39 , 38 , 37 , 39, 38, 35 , 37 , 36 , 35 , 40 ,  
                        40, 36, 38, 39, 42, 42, 36, 36, 35, 41,
                        42, 38, 37, 35, 40, 36, 35, 39, 41, 37,
                        35, 41, 39, 41, 42, 42, 36, 37, 37, 39,
                        42, 35, 36, 41, 41, 41, 39, 39, 35, 39 
                        ],
    "tamaño_arnés" : [58, 58, 52, 58, 57, 52, 55, 53, 49, 54, 
                     59, 56, 53, 58, 57, 58, 56, 51, 50, 59,
                     59, 59, 55, 50, 55, 52, 53, 54, 61, 56,
                     55, 60, 57, 56, 61, 58, 53, 57, 57, 55,
                     60, 51, 52, 56, 55, 57, 58, 57, 51, 59 
                     ]
}

# Crear un DataFrame con los datos utilizando pandas
dataset = pandas.DataFrame(datos)
# Imprimir el DataFrame
dataset
```

```python
# Importar la biblioteca statsmodels con el alias smf
import statsmodels.formula.api as smf

# Definir la fórmula de regresión lineal simple
formula = "tamaño_arranque ~ tamaño_arnés"
# Crear un modelo de regresión lineal simple utilizando la fórmula y el conjunto de datos
modelo = smf.ols(formula = formula, data = dataset)

# Verificar si el modelo tiene parámetros establecidos
if not hasattr(modelo, "params"):
    print ("Modelo seleccionado pero no tiene parámetros establecidos. ¡Necesitamos entrenarlo!") #Imprimimos el mensaje
```

```python
# Entrenar el modelo de regresión lineal simple
modelo_entrenado = modelo.fit()

# Imprimir los parámetros del modelo entrenado
print ("Se han encontrado los siguientes parámetros del modelo:\n" +
        f"Pendiente de línea: {modelo_entrenado.params[1]}\n"
        f"Intercepción de línea: {modelo_entrenado.params[0]}")
```

```python
# Importar la biblioteca de visualización matplotlib
import matplotlib.pyplot as plt

# Crear un gráfico de dispersión de los datos
plt.scatter(dataset["tamaño_arnés"], dataset["tamaño_arranque"])
# Graficar la línea ajustada por el modelo de regresión lineal simple
plt.plot(dataset["tamaño_arnés"], modelo_entrenado.params[1] * dataset["tamaño_arnés"] + modelo_entrenado.params[0], "r", label="Línea ajustada")

# Etiquetar los ejes x e y
plt.xlabel("tamaño_arnés")
plt.ylabel("tamaño_arranque")
# Mostrar la leyenda para la línea ajustada
plt.legend()
```

```python
# Crear un DataFrame con un nuevo valor para tamaño_arnés
tamaño = pd.DataFrame({"tamaño_arnés": [52.5]})
# Predecir el tamaño_arranque utilizando el modelo entrenado
tamaño_arranque_aproximado = modelo_entrenado.predict(tamaño)

# Imprimir el tamaño de arranque aproximado obtenido
print ("Tamaño de arranque aproximado:")
print (tamaño_arranque_aproximado.iloc[0])
```
